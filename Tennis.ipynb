```python
from os import path
from matplotlib import pyplot as plt
import pickle
import numpy as np
```

```python
file_location = path.join('results', 'results_7', 'scores.pickle')
print(file_location)

with open(file_location, 'rb') as f:
    results = pickle.load(f)

results = results.get('average_scores')
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
plt.plot(np.arange(len(results)), results)
plt.ylabel('Average Score Across 100 Episodes')
plt.xlabel('Episode #')
plt.show()
```
